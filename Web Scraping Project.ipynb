```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
#print(soup)
```

Now comes the inspection part of webscraping, we get to the page we want to dowload the data from and understand wt exactly is the info we want to pull and try to figure out how to pull that data.

```python
#As we observed multiple tables are under table tag, lets use indexing to obtain the table we need
soup.find_all('table')[0]
```

```python
table = soup.find_all('table', class_ = 'wikitable sortable')

# find_all= Used to find all the related tags
#find = Used to find only the first instance of that tag we are searching for.
```

```python
table = soup.find_all('table')[0]
#print(table)
```

```python
soup.find_all('th')
# we found that all the table headers are in the th tags, so we tried to check those.
```

```python
# Now lets extract these headers
world_titles = table.find_all('th')
```

```python
#world_titles
```

Now as we have the table headers, lets use list comprehension and pull all these headers into a table

```python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
```

```python
world_table_titles = world_table_titles[1:9]
world_table_titles
```

We now have the table headers, so lets create an pandas dataframe and insert values into respective headers

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

Now lets pull the data into these headers.

```python
# We observed from the inspections that all the data is under tr and td tags,
#tr tag refers to rows and td tags refers to data
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    row_data= row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
#Above we can see that we have two empty lists at the top, unless we get rid of that we will not be append the values into the fields.
# Lets use indexing and remove them.
for row in column_data[2:]:
    row_data= row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)

# That looks good, now lets Append these into there respective positions using the length of the table and loc funtion.
```

```python
all_rows = []
for row in column_data[2:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    all_rows.append(individual_row_data)

df = pd.DataFrame(all_rows, columns=world_table_titles)
```
